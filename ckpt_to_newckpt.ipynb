```python
import torch
# Path to original and updated checkpoint
ckpt_path = "/data/nxtaimraid02/mousakha/repos/orbis/logs/2025-05-20T16-27-30_vq_IF_dino2_e16_DLC11518833_dec/checkpoints/last2.ckpt"
new_ckpt_path = "/data/nxtaimraid02/mousakha/repos/orbis/logs/2025-05-20T16-27-30_vq_IF_dino2_e16_DLC11518833_dec/checkpoints/last.ckpt"
```

```python
#### Tokenizer ####
def remap_key(k: str) -> str:
    # Specific renaming
    k = k.replace("encoder.", "encoder.encoder.")
    k = k.replace("encoder2.", "encoder2.encoder.")
    k = k.replace("post_quant_conv_dino", "post_quant_conv_distill")
    k = k.replace("dino.", "distill.")

    return k

# Load checkpoint
ckpt = torch.load(ckpt_path, map_location="cpu")

# Remap keys
new_state_dict = {}
for k, v in ckpt["state_dict"].items():
    new_k = remap_key(k)
    if new_k:
        new_state_dict[new_k] = v

ckpt["state_dict"] = new_state_dict
torch.save(ckpt, new_ckpt_path)
print(f"✅ Checkpoint saved to: {new_ckpt_path}")
```
